prob2:
plot the score comparision of 2 students in 5 different subjects subjects as xaxis and marks as yaxis

```python
s1=[20,21,18,12,10]
s2=[11,31,52,67,45]
sub=["maths","english","telugu","social","science"]
plt.plot(sub,s1,label="s1",color="cyan")
plt.plot(sub,s2,label="s2",color="orange")
plt.title("marks of students")
plt.legend()
plt.show()
```

```python
s1=[20,21,18,12,10]
s2=[11,31,52,67,45]
sub=["maths","english","telugu","social","science"]
plt.subplot(2,2,1)
plt.plot(sub,s1,label="s1",color="cyan")
plt.legend()
plt.subplot(2,2,2)
plt.plot(sub,s2,label="s2",color="orange")
plt.legend()
plt.show()
```

```python
A=np.array([230,560,780,127,128])
B=np.array([200,160,270,127,400])
print("revenue of company A",A)
print("revenue of company B",B)
profA=np.diff(A)
print("profits of A",profA)
profB=np.diff(B)
print("profits of B",profB)
x=["19-20","20-21","21-22","22-23"]
plt.subplot(1,2,1)
plt.bar(x,profA,color="cyan",label="profit A")
plt.legend(loc="best")
plt.subplot(1,2,2)
plt.bar(x,profB,color="orange",label="profit B")
plt.legend(loc="best")



```
